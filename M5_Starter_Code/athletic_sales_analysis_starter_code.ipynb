```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020 = pd.read_csv("athletic_sales_2020.csv")
sales_2021 = pd.read_csv("athletic_sales_2021.csv")

```

```python
# Display the 2020 sales DataFrame
sales_2020
```

```python
# Display the 2021 sales DataFrame
sales_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020.info()
```

```python
# Check the 2021 sales data types.
sales_2021.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_athletic_sales = pd.concat([sales_2020, sales_2021], axis = 0, join='inner').reset_index(drop = True)
joined_athletic_sales
```

```python
# Check if any values are null.
joined_athletic_sales.isnull().sum()
```

```python
# Check the data type of each column
joined_athletic_sales.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_athletic_sales["invoice_date"] = pd.to_datetime(joined_athletic_sales["invoice_date"])
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_athletic_sales.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_products_sold= joined_athletic_sales.groupby(["region", "state", "city"])["units_sold"].agg(Total_Products_Sold = ("sum"))
total_products_sold.head()


# Show the top 5 results.
total_products_sold.sort_values(by=["Total_Products_Sold"]).head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_sold = pd.pivot_table(joined_athletic_sales, index=["region", "state", "city"], values="units_sold", aggfunc="sum")

total_products2 = total_products_sold.rename(columns={"units_sold": "Total_Products_Sold"})




# Rename the "units_sold" column to "Total_Products_Sold"

total_products2.sort_values(by=("Total_Products_Sold"), ascending=True)



```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = joined_athletic_sales.groupby(["region", "state", "city"])["total_sales"].agg(Total_Sales  = ("sum"))

# Show the top 5 results.
total_sales.sort_values(by=["Total_Sales"], ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
joined_sales_Total = pd.pivot_table(joined_athletic_sales, index=["region", "state", "city"], values="total_sales", aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales3 = joined_sales_Total.rename(columns={"total_sales": "Total_Sales"})

# Show the top 5 results.
total_sales3.head()
```
